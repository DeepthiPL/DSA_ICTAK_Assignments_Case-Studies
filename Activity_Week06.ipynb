# 1. Find out how many countries are there in the dataset.

```python
data['country'].count()
```

# 2. Find out how many countries with respect to the region are available in the dataset

```python
data.groupby('Region', as_index=False).country.count()
```

# 3. On average which region has the highest

#### a. GDP per capita (current US$)

```python
p=data.groupby('Region', as_index=False)[['GDP per capita (current US$)']].mean()
q=p.sort_values(by=['GDP per capita (current US$)'], ascending=False)
q.head(1)
```

#### On average, the region that has the highest GDP per capita, is NorthernAmerica (57659.0).

#### b. International trade: Imports (million US$)

```python
r=data.groupby('Region', as_index=False)[['International trade: Imports (million US$)']].mean()
s=r.sort_values(by=['International trade: Imports (million US$)'], ascending=False)
s.head(1)
```

#### On average, the region that has the highest International trade: Imports, is NorthernAmerica (663552.25).

# 4. Find out which region has more consistent with respect to GDP per capita (current US$)

```python
t=data.groupby('Region', as_index=False)[['GDP per capita (current US$)']].std()
u=t.sort_values(by=['GDP per capita (current US$)'])
u.head(1)
```

#### The region which has more consistent GDP per capita, is WesternAfrica (815.095399).
