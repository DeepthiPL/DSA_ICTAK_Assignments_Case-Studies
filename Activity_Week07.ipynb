# 3. Examine the variables ‘Hardness’, ‘Sulfate’, and ‘Organic_carbon’ and comment on your observations.

```python
data1=data[['Hardness', 'Sulfate', 'Organic_carbon']]
corrmatrix=data1.corr()
```

```python
plt.subplots(figsize=(20,8))
sns.heatmap(corrmatrix, vmin=-1, vmax=1, annot=True,linewidths=0.2, cmap='coolwarm')
```

# Insights :

## The correlation values for ‘Hardness’, ‘Sulfate’, and ‘Organic_carbon' are as follows:
#### Hardness vs Sulfate : -0.11
#### Hardness vs Organic_carbon : 0.013
#### Sufate vs Organic_carbon : 0.027

## Hence, it can be inferred as follows : 
#### Hardness and Sulfate shows a strong negative correlation. That means, as the amount of Sulfate increases hardness decreases.
#### Hardness vs Organic_carbon and Sufate vs Organic_carbon shows positive correlation. That means, as the amount of Organic_carbon increases, Hardness increases. Also, as the amount of Sulfate increases, the amount of Organic_carbon increases. However, the correlation value is very small.

#### ph and Chloramines shows a strong negative correlation and that means as the amount of ph increases, the amount of Chloramines decreases. Whereas, ph and Conductivity shows a weak positive correlation and that means as the amount of ph increases, the Conductivity also increases, but only slightly.
