```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_data=pd.read_csv('iris.csv')
iris_data
```

# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

# Visualization 1

```python
plt.figure(figsize=(6,4))
plt.scatter(iris_data['SL'], iris_data['Classification'], s=10, c='red', marker='^')
plt.title('Plot of SL vs Classification', fontsize=16)
plt.xlabel('SL')
plt.ylabel('Classification')
```

# VIsualization 2

```python
sns.pairplot(iris_data)
```

# Visualization 3

```python
sns.boxplot(x='SW', y='Classification', data=iris_data)
plt.title('Boxplot of SW', fontsize=16)
```

# Visualization 4

```python
sns.heatmap(iris_data.corr(), annot=True)
plt.xticks(rotation=45)
```

# Visualization 5

```python
sns.violinplot(x='PL', y='Classification', data=iris_data)
```
